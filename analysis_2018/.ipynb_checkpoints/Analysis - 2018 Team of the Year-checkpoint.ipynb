# NRL Team of the Year Analysis

The Dally M Team of the Year is the National Rugby League of Australia's equivalent to an all-Pro team in American sports.  Each week, journalists and commentators across Australia pick their team of the week and at the end of the season, the votes are combined to select a Team of the Year.  Since most of the people voting on team of the week are doing so subjectively, there is often a bit controversy when the Dally M Team of the Year is revealed at the end of the season.  Last year is a prime example:

https://www.sportingnews.com/au/league/news/dally-m-awards-fans-blow-up-over-dally-m-team-of-the-year-selections/1mds08rg0caso1afr28idjyzyo

While most of the 9 selections in 2018's Team of the Year were well received, there was plenty of outrage on social media over a few, most notably the selections of Joseph Leilua (centre) and Josh Jackson (second-row).  


In this analysis I will look into data from the 2018 NRL season to determine whether or not the Dally M Team of the Year was an accurate reflection of player performance.  The data I'm using was scraped from the official NRL website (www.nrl.com) using Selenium and Python's BeautifulSoup library and compiled into a database.  The database contains detailed player stats from every match.  In order to determine whether or not a player deserved to be in Team of the Year, I will first analyze various stats to determine which were the most highly correlated with a team's number of wins over the course of the season.  Once I've determined which stats are most important in contributing to a victory, I will compare players at each position by these stats to determine who deserved to be in 2018's Team of the Year.

```python
#Import Pandas and connect to database
import pandas as pd
import mysql.connector as sql
from dotenv import load_dotenv
import os

load_dotenv()

mydb = sql.connect(
  host="localhost",
  user=os.getenv("DB_USER"),
  passwd=os.getenv("DB_PASSWORD"),
  database="NRL_data"
)

#Function to easily create dataframes from SQL query results
def create_df(query):
    return pd.read_sql_query(query, con=mydb)
```

# 2018 Final Team Standings

In order to determine which stats have the strongest correlation with total wins, I will put together an easy to read final standings list from the 2018 season.  The database contains entries for each individual match but does not aggregate wins or point differential for each team.

```python
# Replace team_ids in Matches table with with team names from Teams table
match_query = '''SELECT m.id, m.round, m.date,
CASE
	WHEN m.home_team_id = m.winner THEN home.nickname
    ELSE away.nickname
END winner,
CASE
	WHEN m.home_team_id = m.winner THEN m.home_score
    ELSE m.away_score
END winning_score,
CASE
	WHEN m.home_team_id = m.winner THEN away.nickname
    ELSE home.nickname
END loser,
CASE
	WHEN m.home_team_id = m.winner THEN m.away_score
    ELSE m.home_score
END losing_score, home.nickname home, away.nickname away
FROM Matches m
JOIN Teams home
ON m.home_team_id = home.id
JOIN Teams away
ON m.away_team_id = away.id
WHERE year(m.date) = 2018
ORDER BY date;'''
matches_df = create_df(match_query).set_index('id')
matches_df.head()
```

```python
#Create dataframe that combines points_scored and points_allowed in each team's victory
winners = matches_df.groupby('winner')
w_scored = winners['winning_score'].sum()
w_allowed = winners['losing_score'].sum()

#Create dataframe that combines points_scored and points_allowed in each team's loss
losers = matches_df.groupby('loser')
l_scored = losers['losing_score'].sum()
l_allowed = losers['winning_score'].sum()

#Combine dataframes
total_scores = pd.concat([w_scored, l_scored, w_allowed, l_allowed], axis=1, join='inner')
total_scores.columns = ['scored_in_win', 'scored_in_loss', 'allowed_in_win', 'allowed_in_loss']
total_scores.head(2)

#Calculate total points_differential, points_scored and points_allowed for each team
total_scores['pts_scored'] = total_scores['scored_in_win'] + total_scores['scored_in_loss']
total_scores['pts_allowed'] = total_scores['allowed_in_win'] + total_scores['allowed_in_loss']
total_scores['pts_differential'] = total_scores['pts_scored'] - total_scores['pts_allowed']
total_points_data = total_scores[['pts_differential', 'pts_scored', 'pts_allowed']]
total_points_data.head()
```

```python
#Aggreate win / loss records for each team
win_loss_query = '''SELECT t.nickname team,
    count(m.winner) as wins,
    24 - count(m.winner) as loses
FROM Matches m
JOIN Teams t
ON m.winner = t.id
WHERE year(m.date) = 2018
GROUP BY m.winner
ORDER BY wins DESC;'''
win_loss_df = create_df(win_loss_query).set_index('team')
win_loss_df.head()
```

```python
#Combine win/loss records with point differentials to create final standings
final_standings = pd.concat([win_loss_df, total_points_data], axis=1, join='inner').sort_values(['wins', 'pts_differential'], ascending=[0,0])
final_standings
```

```python
#Visualize if points_scored correlates highly with total wins
final_standings.plot.scatter(x='wins', y='pts_scored', color='Blue', label='scored')
```

We can see that there is a very weak correlation betwen points_scored and wins. Lets look at alternative metrics and player stats to find a better correlations with win totals.

# Finding correlations between win total and various team stats

The PlayerMatchStats table contains one entry for each player involved in every match.  I will combine each player's individual contribution into one table that aggregates each statistical category by team for the 2018 season

```python
#Replace player_ids and team_ids with names in the PlayerMatchStats table
query = '''SELECT CONCAT(p.first_name, ' ' ,p.last_name) name, t.nickname team, p_stats.*
        FROM PlayerMatchStats p_stats
        JOIN Players p
        ON p_stats.player_id = p.id
        JOIN Teams t
        ON p_stats.team_id = t.id
        JOIN Matches m
        ON p_stats.match_id = m.id
        WHERE year(m.date) = 2018;'''

player_stat_df = create_df(query)

player_stats_grouped = player_stat_df.groupby(['name', 'team']).agg(sum)
player_stats_grouped.head()
```

In order to make this analysis a little more manageable, I will remove some of the stats that refer to events which rarely occur over the course of a match.

```python
player_stat_df.columns
```

From the list of columns it looks like almost all of the stats fall into one of several categories: scoring, running, kicking, defensive/tackling, and errors/penalties. To simplify the anlaysis, I will only include the more encompassing stats for each category. Using running stats as an example, I will include total_run_metres but ignore more refined stats like kick_return_metres and post_contact_metres which both are included in total_run_metres.  I will also create columns for stats like metres_per_kick and tackle_percentage which aren't included in the initial PlayerMatchStat table but can easily be calculated.

```python
team_stats = player_stat_df.groupby('team').agg(sum)

#Combine aggregated stats with teams' win total
slimmed_team_stats = team_stats[['points', 'tries', 'try_assists', 'total_runs', 'total_run_metres', 'tackles_made', 'tackles_missed', 'ineffective_tackles',
           'kicks', 'kicking_metres', 'errors', 'penalties']]
wins = final_standings[['wins', 'pts_differential', 'pts_scored', 'pts_allowed']]
team_stats = pd.concat([wins, slimmed_team_stats], axis=1, join='inner').sort_values(['wins', 'pts_differential'], ascending=[0,0])

#Add columns for categories that calculate averages
team_stats['metres_per_kick'] = team_stats['kicking_metres'] / team_stats['kicks']
team_stats['metres_per_run'] = team_stats['total_run_metres'] / team_stats['total_runs']
team_stats['tackle_percentage'] = team_stats['tackles_made'] / (team_stats[['tackles_made', 'tackles_missed', 'ineffective_tackles']].sum(axis = 1))

team_stats
```

```python
#Visualize correlations between different statistical categories
corr = team_stats.corr()
corr.style.background_gradient(cmap='coolwarm')
```

We can see from this chart how various stats in our analysis correlate with a team's win total. There are some fairly obvious stats with strong correlations like points_scored, total_run_metres and tries (the rugby equivalent of a touchdown in football). There are also some stats with correlations that are a bit surprising.  I would have predicted to see a strong negative correlation between wins and errors, however, the correlation is slightly positive although very weak.  This could be explained by the fact that many errors fall under the category of "handling errors" which means that a team with posession dropped the ball, thereby surrending it to the other team.  Teams with more position are more likely to commit handling errors and similarly, teams with more posession are likely to score more points which correlates highly with winning.

# Identifying top performers at each position

I will now compare players at each position based off of the stats that have the biggest impact on a team's win total. When conducting deeper analysis on the controversial selections, I will also take into conserdation the stats that are more applicable to the individual positions.  Player positions in rugby can be fall under the categories of forwards or backs. Forwards tend to be larger, slower players who have more of a defenisve impact while spending most of their time towards the middle of the field.  Backs tend to play out wide because they're smaller and quicker which makes them more of an attacking threat in space.  Backs also tend to do the vast majority of a team's kicking. 

For the purposes of this analyis, each player will be assigned to the position at which they played the most matches over the course of the 2018 season

```python
#Find position at which each player played the most games
from scipy.stats import mode

position_df = player_stat_df[['name', 'position_id']]
modes = position_df.groupby('name')['position_id'].apply(lambda x: mode(x)[0][0]).reset_index().rename(columns={"position_id": "main_position"})

#Combine player positions with dataframe of aggregate statistics
player_stats_final = pd.merge(modes, player_stats_grouped, how='left', on='name')

#Create columns for additional statistics
player_stats_final = player_stats_final.drop(columns=['match_id', 'position_id'])
player_stats_final['metres_per_run'] = player_stats_final['total_run_metres'] / player_stats_final['total_runs']
player_stats_final['metres_per_kick'] = player_stats_final['kicking_metres'] / player_stats_final['kicks']
player_stats_final['tackle_percentage'] = player_stats_final['tackles_made'] / (player_stats_final['tackles_made'] + player_stats_final['tackles_missed'] + player_stats_final['ineffective_tackles'])

#Clean up dataframe
player_stats_final = player_stats_final.drop(columns=['id', 'player_id', 'team_id'])
stat_columns = player_stats_final.columns[2:]
player_stats_final = player_stats_final.groupby(['name', 'main_position'])[stat_columns].agg('sum')
player_stats_final = player_stats_final.reset_index()
player_stats_final.set_index('name').head()
```

Now that we have a finalized dataframe displaying all players, their main position, and their stats from the course of the 2018 season, we can analyze who stood out at each position.  Since we're looking for top performers over the course of a season, I will remove all players who played less than 960 minutes (the equivalent of 12 full 80-minute matches, which is equal to half of the season) from this analysis.

```python
#Remove everyone who played less than the equivalent of 10 full matches 
p_stats_filtered_minutes = player_stats_final[player_stats_final['minutes_played'] > 960]

#Create dictionary of player stats for each position
main_stats = ['name', 'main_position', 'minutes_played', 'tries', 'try_assists', 'total_run_metres', 'total_runs', 'metres_per_run', 'kicks', 'kicking_metres', 'metres_per_kick', 'tackles_made', 'tackle_percentage']
positions = ['Fullback', 'Wing', 'Centre', 'Five-Eigth', 'Halfback', 'Prop', 'Hooker', '2nd Row', 'Lock']
positions_dict = {}
for i in range(10):
    if i > 0:
        p_stats_filtered_minutes = p_stats_filtered_minutes[main_stats]
        position = p_stats_filtered_minutes[p_stats_filtered_minutes['main_position'] == i]
        positions_dict[i] = position.sort_values(['total_run_metres', 'tries'], ascending = [0,0])
        
#Convert dictionary into a dataframe and only include top performers at each position 
stat_leader_df = pd.DataFrame()      
for key in positions_dict:
    stat_leader_df = pd.concat([stat_leader_df, positions_dict[key][:7]])

stat_leader_df['main_position'] = stat_leader_df['main_position'].apply(lambda x: positions[x - 1])
stat_leader_df.set_index('name')
```

In the above analysis, I chose to sort players first by total_run_metres and then by tries.  Even though tries have a stronger correlation with wins than total running metres, many tries involve the contributions of many different players similar to touchdowns in football.  In order to determine which individual players excelled in comparison to others, I felt that running meters would be a good starting point since this is one of the strongest indicators of a player's overall performance regardless of their position or stle of play.

# Comparing the 2018 Team of the Year to stat leaders

## Centres

One of the most controversial selections in the 2018 Team of the Year was Joseph Leilua of the Canberra Raiders at Centre.  Many felt that Latrell Mitchell of the Sydney Roosters should have been selected.  Players who are Centres play out wide and tend to be a team's most powerful ball-runners. For this reason, running stats and tries are the best measures of a players performance. Centres aren't as involved defensively as other positions, however, missed tackles at this position can be costly and more likely to lead to a try to the opposition. For this reason, we don't necessarily care if a centre has a high tackle count, however, we'd like them to have a high tackle percentage.

```python
#Do a deep analysis into running stats
centre_df = p_stats_filtered_minutes[p_stats_filtered_minutes['main_position'] == 3]
centre_df = centre_df[main_stats].drop(columns=['kicks', 'main_position'])
centre_df = centre_df.sort_values(['total_run_metres', 'tries', 'metres_per_run', 'tackle_percentage'], ascending = [0,0,0,0]).set_index('name')
centre_df.head()
```

I first generated a dataframe of all centres who played more than 10 full matches and sorted them by total_run metres. Surprisingly neither Joseph Leilua or Latrell Mitchell make the top 5. Lets compare Leilua and Mitchell to a few of the top ball-runners.

```python
centre_list = ['Will Hopoate', 'Esan Marsters', 'Euan Aitken', 'Joseph Leilua', 'Latrell Mitchell']
top_centres = centre_df.loc[centre_list]
top_centres
```

Now we can see why Leilua and Mitchell were viewed more favorably than some of the players who had more rushing metres.  Both Leilua and Mitchell had a much stronger impact in scoring or setting up tries for their team. 

The Sporting News article mentions that the Team of the Year is selected by combining 'team of the week' votes. Each week, rugby journalists and commentators pick their top 3 players at each position. First place gets 3 votes, second place gets 2 votes, and third place gets 1 votes.  I will simulate this voting process to see if this influenced Joseph Leilua's selection.  Its possible that fans viewed Mitchell more favorably because he was more consistant week-after-week, while Leilua was inconsistent but ended up in the top three during his good weeks. 

In order to rank players on a week-by-week basis, I will create a formula that gives a player a score based off of their weekly stats.  This score will be calculated similarly to a weekly points total in fantasy football.  This formula can also be updated by position.

```python
pd.options.mode.chained_assignment = None

#Add match-round to player_stat_df
rounds = matches_df['round']
stats_by_round = player_stat_df.join(rounds, on='match_id', lsuffix='_player', rsuffix='_match').set_index('name')

#Formula for rating a player's weekly performance
def centre_rating_formula(df):
    return ((df['tries'] * 5) + (df['total_run_metres'] / 10) + 
            (df['try_assists'] * 2.5) + df['steals'] + (df['tackle_percentage'].astype(float) * 10) +
            df['intercepts'] - (df['penalties']) - df['errors'])

#Function to determine top performers of the week at each position
def find_top_players(position_id, rating_formula):
    p_stats = stats_by_round[stats_by_round['position_id'] == position_id]
    top_performers_by_round = {}
    for i in range(1, 26):
        round_df = p_stats[p_stats['round'] == i]
        round_df['rating'] = rating_formula(round_df)
        round_df = round_df.sort_values('rating', ascending=0).head(3).reset_index()
        round_df['round_score'] = [3, 2, 1]
        top_performers_by_round[i] = round_df[['name', 'round_score']]
    return top_performers_by_round

top_centres = find_top_players(3, centre_rating_formula)

#We want to visualize changes in players' cumulative 'team of the week' votes
import matplotlib.pyplot as plt

#First create a dictionary with the names of centres we want to compare as the keys
weekly_tally = dict.fromkeys(centre_list)
#For each centre, create another dictionary that shows their number of votes at the end of each round
for player in centre_list:
    weekly_tally[player] = {}
    for i in range(1, 26):
        weekly_score_df = top_centres[i].set_index('name')
        if i == 1:
            if player in list(weekly_score_df.index.values):
                weekly_tally[player][i] = weekly_score_df.loc[player, 'round_score']
            else:
                weekly_tally[player][i] = 0
        else:
            if player in weekly_score_df.index.values:
                weekly_tally[player][i] = weekly_tally[player][i - 1] + weekly_score_df.loc[player,'round_score']
            else:
                weekly_tally[player][i] = weekly_tally[player][i - 1]

top_centres
#Plot the results
fig = plt.figure(figsize=(15, 7))
for player in list(weekly_tally.keys()):
    x = list(range(len(weekly_tally[player])))
    y = list(weekly_tally[player].values())
    plt.plot(x, y, label=player)
plt.legend(loc='upper left')
plt.xlabel("Round Number")
plt.ylabel("Cumulative Ranking")
plt.show()
```

Despite the harsh criticism from social media, we can see that Joseph Leilua's Team of the Year selection is justified given the voting process.  Despite not even finishing in the top 7 players for rushing metres, Latrell Mitchell finished tied for second place, most likley due to his huge impacting in scoring and setting up tries.  Although one can easily end up with different results by playing around with the formula for players of the week, we can conclude that the current formula is pretty accurate in giving weight to the statistics that matter to rugby fans and journalists.

## Second Row

Another controversial selection in the 2018 Team of the Year was Josh Jackson at Second Row.  In order to judge whether or not this was a fair selection based off the Team of the Year selection process,  I will update the scoring formula that calculates top performers by week so it more accurately grades players who are in the Second Row.  Second Rowers are forwards who are larger and play towards the center of the field. While we would like to see these players get a lot of running metres and score tries, we primarily want these players to have lots of tackles. Their main priority is to make as many tackles as possible, trying to stop the opposition as quickly as possible before they have the chance to get the ball out wide to their major try-scoring threats.

```python
#Update weekly performance formula to place more weight on tackles and slightly less weight on tries and try_assists
def sr_rating_formula(df):
    return ((df['tries'] * 4) + (df['total_run_metres'] / 10) + 
        (df['try_assists'] * 2) + df['steals'] + (df['tackle_percentage'].astype(float) * 5) +
        (df['tackles_made'] * .5) + df['intercepts'] - (df['penalties']) - df['errors'])

#Put together dataframe of top Second Rowers
sr_df = p_stats_filtered_minutes[p_stats_filtered_minutes['main_position'] == 8]
sr_df = sr_df[main_stats].drop(columns=['kicks', 'main_position'])
sr_df = sr_df.sort_values(['total_run_metres', 'tries', 'total_runs', 'tackle_percentage'], ascending = [0,0,0,0])
sr_df = sr_df.set_index('name').head(15)

#This is the function we used previously to determine top performers each week
top_srs = find_top_players(8, sr_rating_formula)

#Tally up the votes for each player from the 15 man list we created above
weekly_tally = dict.fromkeys(sr_df.index.values)
for player in list(sr_df.index.values):
    weekly_tally[player] = {}
    for i in range(1, 26):
        weekly_score_df = top_srs[i].set_index('name')
        if i == 1:
            if player in list(weekly_score_df.index.values):
                #score = weekly_score_df[player]['round_score']
                weekly_tally[player][i] = weekly_score_df.loc[player]['round_score']
            else:
                weekly_tally[player][i] = 0
        else:
            if player in weekly_score_df.index.values:
                weekly_tally[player][i] = weekly_tally[player][i - 1] + weekly_score_df.loc[player]['round_score']
            else:
                weekly_tally[player][i] = weekly_tally[player][i - 1]

#Create a dictionary displaying each player with their final vote count at the end of the season
final_votes_count = {}
for player in weekly_tally.keys():
    final_votes_count[player] = weekly_tally[player][25]
sorted_vote_counts = sorted(final_votes_count.items(), key=lambda x: x[1], reverse=True)

#Limit this dictionary the 4 players with the highest scores
top_srs = [i[0] for i in sorted_vote_counts[:4]]

#Get week-by-week data for the 4 players with the highest scores
top_srs_dict = {}
for player in weekly_tally.keys():
    if player in top_srs:
        top_srs_dict[player] = weekly_tally[player]

#Plot the vote totals of the top 4 players over the course of the season
fig = plt.figure(figsize=(15, 7))
for player in list(top_srs_dict.keys()):
    x = list(range(len(top_srs_dict[player])))
    y = list(top_srs_dict[player].values())
    plt.plot(x, y, label=player)
plt.legend(loc='upper left')
plt.xlabel("Round Number")
plt.ylabel("Cumulative Ranking")
plt.show()
```

```python
top_srs = sr_df.loc[['Angus Crichton', 'John Sutton', 'Boyd Cordner', 'Josh Jackson']]
top_srs
```

Once again, we can see how the Team of the Year voting process justifies another controversial selection.  Josh Jackson didn't finish in the top 10 for running meters and didn't stand out in regards to try scoring, however, he was able to register enough weekly top-3 finishes with our formula to be tied for Team of the Year selection.    

We can also see that whether consciously or unconsciously, voters used a different set of criteria to judge players at different positions.  When using the centre's formula to determine the best Second Rowers on a week-by-week basis, Josh Jackson didn't finish in the top-10.  However, when less weight was giving to try-scoring statistics and an emphasis was placed on total tackles, Josh Jackson finished tied for 1st mostly due to his huge tackle count over the course of the season.

# Fullbacks

When I initially created the "stat_leader_df" at the beginning of this analysis which ranked the top players at each position, the position that intrigued me the most was Fullback.  Roger Tuivasa-Sheck of the Warriors was selected to Team of the Year and also won the Dally M Player of the Year award, which is equivalent to MVP in American sports.  This selection was almost univerally accepted by rugby supporters, however, the stats seem to indicate that James Tedesco of the Sydney Roosters outplayed Tuivasa-Sheck in almost every important statistical category.

```python
stat_leader_df.set_index('name').loc[['James Tedesco', 'Roger Tuivasa-sheck']]
```

We can see that Tedesco and Tuivasa-Sheck played almost the same exact number of minutes, yet Tedesco had more running metres, metres per run, tackles, a better tackle percentage, and most importantly was involved in significantly more tries.  

```python
top_fbs = find_top_players(1, centre_rating_formula)
points_tally = dict.fromkeys(['James Tedesco', 'Roger Tuivasa-sheck'], 0)
for round in list(top_fbs.keys()):
    for fb in ['James Tedesco', 'Roger Tuivasa-sheck']:
        try:
            points = top_fbs[round].set_index('name').loc[fb, 'round_score']
            points_tally[fb] += points
        except:
            continue
points_tally
```

We can also see that over the course of the season, Tedesco was consistently a top performer at his position more often than Tuivasa-Sheck by receiving 26 weekly top-3 votes, compared to Tuivasa-Sheck's 11 votes.  Since fullbacks are similar to centres in terms physical attributes and desired skills, I used the rating formula that had already been created for the Centre analysis.

### Rank players based on how big an impact they had for their respective teams

One possible explanation for Tuivasa-Sheck's inclusion in Team of the Year is that he played for a worse team than Tedesco.  Tedesco plays for the Sydney Roosters who boast a line-up full of stars and who ultimately won the 2018 Premership easily beating everyone they faced once the playoffs began.  Despite only having one less regular season win, the Warriors had a much lower points differential than the Roosters and were viewed as the surprise team of the season who many originally predicted to finish dead-last.

Its possible that much of the Warriors success was due to Tuivasa-Sheck's strong performances while Tedesco benefited from playing for a stronger team.  To investigate, I will analyze how much of an impact each player had for their respective team at each of the major statistical categories.  If this theory is correct, we should see that Tuivasa-Sheck was responsible for a larger percentage of the Warriors statistical output than Tedesco was for the Roosters.

```python
#Formula takes in dictionary in format {player: team} and returns player impact on team output
def compare_individual_contribution_to_team(player_dict):
    def find_team_stats(team_name, stat):
        stats = team_stats.loc[team_name]
        return stats[stat]
    
    individual_impact = {}

    for name, team in player_dict.items():
        individual_impact[name] = {}
        individual_impact[name]['name'] = name
        player_df = stat_leader_df[stat_leader_df['name'] == name]
        player_df = player_df[['name', 'tries', 'try_assists', 'total_run_metres', 'total_runs', 'kicks', 'kicking_metres', 'tackles_made']]
        i = 0
        while i < len(player_df.columns):
            try:
                combined_team_stat = find_team_stats(team, player_df.columns[i])
                percent_of_team_stat = player_df[player_df.columns[i]] / combined_team_stat
                individual_impact[name][player_df.columns[i]] = (percent_of_team_stat.item() * 100)
                i += 1
            except:
                i += 1
                continue

    comparison_dataframe = pd.DataFrame()
    for i in individual_impact:
        player_dataframe = pd.DataFrame([individual_impact[i]], columns=individual_impact[i].keys())
        comparison_dataframe = pd.concat([comparison_dataframe, player_dataframe])
    return comparison_dataframe.set_index('name').round(2)
```

```python
fullbacks = {'James Tedesco': 'Roosters', 'Roger Tuivasa-sheck': 'Warriors'}
fb_comparison = compare_individual_contribution_to_team(fullbacks)

fb_bar_graph = pd.DataFrame({'James Tedesco': fb_comparison.loc['James Tedesco'],
                      'Roger Tuivasa-Sheck': fb_comparison.loc['Roger Tuivasa-sheck']},
                     index = list(fb_comparison.columns))

import matplotlib.ticker as mtick
ax = fb_bar_graph.plot(kind='bar', figsize=(15, 7), legend=True, fontsize=12, rot=0)
ax.legend(prop={'size': 15})
ax.set_xlabel("Stat Category", fontsize=15)
ax.set_ylabel("Impact", fontsize=15)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())

plt.show()
```

This bar graph shows us that James Tedesco made a larger contribution to his team's statistical output over the course of 2018 than Roger Tuivasa-Sheck.  Combining this data with Tedesco's superior fullback of the week vote count, its very surprising that Tuivasa-Sheck was named to the Team of the Year over Tedesco.

From this analysis, we can see that the collective opinion of fans should not take the place of statistical analysis when evaluating athletic performance.  We demonstrated that two of the more controversial selections for the NRL's 2018 Team of the Year were completely justified given the selection criteria and demands of the individual positions, while one of the most widely agreed upon selections appears to have been misguided.
